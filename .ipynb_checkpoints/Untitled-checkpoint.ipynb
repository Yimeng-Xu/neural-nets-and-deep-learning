### 1.1 Lasso

```python
clf = linear_model.Lasso(alpha = 0.8, normalize = True, fit_intercept = False)
clf.fit(hitters_new.iloc[:,:-1], hitters_new.iloc[:,-1:])
clf.get_params(deep = True)
```
