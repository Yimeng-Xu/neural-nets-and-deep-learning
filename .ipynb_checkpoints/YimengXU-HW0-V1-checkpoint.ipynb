# Question 1

```python
import pandas as pd
import numpy as np
from sklearn import linear_model
from sklearn.linear_model import lasso_path, LassoCV, Ridge, RidgeCV
from itertools import cycle
import matplotlib.pyplot as plt
# Change figure size once and for all
from pylab import rcParams
rcParams['figure.figsize'] = 10,6
```

```python
pd.set_option('display.max_rows', 800)
```

```python
# Read in the data
hitters = pd.read_csv("Hitters.csv")
```

```python
hitters.head(5)
```

```python
hitters.columns
```

```python
# Name the "Unnamed" column
hitters.rename(columns = {'Unnamed: 0': 'Name'}, inplace = True)
```

```python
hitters.columns
```

```python
# Omit categorical predictors
hitters_new = hitters._get_numeric_data()
```

```python
hitters_new.columns
```

```python
# Find out the number of missing values in each variables in the dataset
hitters_new.isnull().sum(axis = 0)
```

```python
# Fill the missing values in the salary column with the column mean
hitters_n = hitters_new.fillna(hitters_new.mean())
```

```python
hitters_n.isnull().sum()
```

```python
hitters_n.shape
```

```python
# Separate the predictors and the target variable
X = hitters_n.iloc[:,:-1]
y = hitters_n.iloc[:,-1:]
# Normalize the predictors
X /= X.std(axis = 0)
```

## 1.1 Lasso Regression

```python
# A first try on fitting the model
clf = linear_model.Lasso(alpha = 2, normalize = True, fit_intercept = False, max_iter = 10000)
clf.fit(X, y)
# Print out the coefficients
print("The coefficients of the model with regularization penalty = 2:")
for col, coef_l in zip(hitters_n.columns, clf.coef_):
        print(col, coef_l)

# Predict salary of each player using the model and the original predictors
SalaryPred = clf.predict(X)
SalaryPred = pd.DataFrame(SalaryPred)
SalaryPred.rename(columns = {0:'SalaryPred'}, inplace = True)
hitters_pred = pd.concat([hitters_n, SalaryPred], axis=1)
hitters_pred[['Salary', 'SalaryPred']].head()
```

```python
# Coefficient trajectories, aka lasso path
eps = 5e-5
alphas_lasso, coefs_lasso, _ = lasso_path(X, y, eps, fit_intercept = False)

# Display results
plt.figure(num = 1)
colors = ['#000000', '#87CEFA', '#FFB6C1', '#AB82FF', '#7CCD7C', '#FA8072', '#EE4000', '#7FFF00', 
          '#800000', '#0000FF', '#FFC0CB', '#FF7F24', '#FFB90F', '#FFB90F', '#C0FF3E', '#FFFF00']
lines = cycle([':', '-.', '--', '-'])

for i,(coef_l, c, line) in enumerate(zip(coefs_lasso[0], colors, lines)):
    l1 = plt.plot(alphas_lasso, coef_l, c=c, linestyle = line, label = hitters_n.columns[i])

plt.xlabel('alpha')
plt.ylabel('coefficients')
plt.title('Coefficient Trajectory')
plt.axis('tight')
plt.legend(loc = 'lower right', ncol = 2)
plt.show()
```

#### The final three predictors that remain in the model are "AtBat", "Years" and "Walks".

### Select the optimal regularization penalty using cross validation

**We adopt LOOCV since the dataset is small and it won't be computationally expensive.**

```python
model = LassoCV(cv=322, fit_intercept = False).fit(X, y.values.ravel())
# Display results
ymin, ymax = 0, 400000
plt.plot(model.alphas_, model.mse_path_, ':')
plt.plot(model.alphas_, model.mse_path_.mean(axis=-1), 'k',
         label='Average across the folds', linewidth=2)
plt.axvline(model.alpha_, linestyle='--', color='k',
            label='Optimal alpha: CV estimate')

plt.legend()

plt.xlabel('alpha')
plt.ylabel('Mean square error')
plt.title('Mean square error on each fold')
plt.axis('tight')
plt.ylim(ymin, ymax)
plt.show()
print("The optimal regularization penalty:", model.alpha_)
```

**The optimal regularization penalty is around 1.47.**

```python
# Find out the number of predictors in that model
# Print out the coefficients
print("The coefficients of the model with the optimal regularization penalty:")
for col, coef_l in zip(hitters_n.columns, model.coef_):
        print(col, coef_l)
```

**There are 11 predictors in the model with the optimal regularization penalty estimate by cross validation.**

**Predict salaries using Lasso regression:**

```python
# Predict salary of each player using the Lasso model with the optimal regularization and the original predictor data
SalaryPred_l = model.predict(X)
SalaryPred_l = pd.DataFrame(SalaryPred_l)
SalaryPred_l.rename(columns = {0:'SalaryPred'}, inplace = True)
hitters_pred_l = pd.concat([hitters_n, SalaryPred_l], axis=1)
hitters_pred_l[['Salary', 'SalaryPred']].head()
```

## 1.2 Ridge Regression

```python
# The coefficient trajectory
clf_r = Ridge()
coefs = []
errors = []
alphas = np.logspace(-6, 6, 200)

# Train the model with different regularization penalties
for a in alphas:
    clf_r.set_params(alpha = a)
    clf_r.fit(X, y)
    coefs.append(clf_r.coef_[0])

# Display results
ax = plt.gca()
ax.plot(alphas, coefs)
ax.set_xscale('log')
plt.xlabel('alpha')
plt.ylabel('coefficients')
plt.title('Ridge coefficients as a function of the regularization')
plt.axis('tight')
plt.show()
```

```python
# Use cross validation to find out the optimal regularization value
# For simplicity, we use LOOCV, which is the default CV method of the RidgeCV function
## and only if using this method, we can take advantage of the function's ability to store mean squared errors
model_r = RidgeCV(alphas, fit_intercept = False, store_cv_values = True).fit(X, y.values.ravel())
# Display results
plt.plot(alphas, model_r.cv_values_.T, ':')
plt.plot(alphas, model_r.cv_values_.T.mean(axis=-1), 'k',
         label='Average across the folds', linewidth=2)
plt.axvline(model_r.alpha_, linestyle='--', color='k',
            label='Optimal alpha: CV estimate')

plt.legend()

plt.xlabel('alpha')
plt.ylabel('Mean square error')
plt.title('Mean square error on each fold')
plt.axis('tight')
#plt.ylim(ymin, ymax)
plt.show()
print("The optimal regularization penalty:", model_r.alpha_)
```

### The optimal regularization penalty is 2.146 and the estimated coefficients under this penalty is as follows.

```python
# Find out which predictors are in the model with the optimal regularization penalty
for col, coef_or in zip(hitters_n.columns, model_r.coef_):
    print(col, coef_or)
```

**Using Ridge regression and the optimal regularization estimated by LOOCV, the model predicting salaries contains all the predictors.**

**Predict salaries using Ridge regression:**

```python
# Predict salary of each player using the Ridge model with the optimal regularization and the original predictor data
SalaryPred_r = model_r.predict(X)
SalaryPred_r = pd.DataFrame(SalaryPred_r)
SalaryPred_r.rename(columns = {0:'SalaryPred'}, inplace = True)
hitters_pred_r = pd.concat([hitters_n, SalaryPred_r], axis=1)
hitters_pred_r[['Salary', 'SalaryPred']].head()
```

# Question 2

The bias-variance tradeoff is the dilemma that predictive models with a lower bias tend to have a higher variance of the parameter estimates across different samples.<p>
Models with a higher variance may have more predictors than models with a lower variance and tend to have more curvature. But models with less predictors have a higher bias.<p>
In Question 1, the Lasso regression model with the optimal regularization have only 11 predictors while the Ridge regression with the optimal regularization have all of the 16 predictors. The Lasso model has a lower variance and a higher bias, whereas the Ridge regression has a higher variance and a lower bias.
