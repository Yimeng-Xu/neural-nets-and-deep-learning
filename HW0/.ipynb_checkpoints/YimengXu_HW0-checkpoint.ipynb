# Question 1

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn import linear_model
from sklearn.linear_model import lasso_path, LassoCV, Ridge, RidgeCV
from itertools import cycle
import matplotlib.pyplot as plt
# Change figure size once and for all
from pylab import rcParams
rcParams['figure.figsize'] = 12, 8
```

### Preprocessing data

```python
# Read in the data
hitters = pd.read_csv("Hitters.csv")
#hitters.head(5)

# Name the "Unnamed" column
hitters.rename(columns = {'Unnamed: 0': 'Name'}, inplace = True)
# Omit categorical predictors
hitters_new = hitters._get_numeric_data()
print("The categorical predictors that are omitted:")
print([col for col in hitters.columns if col not in hitters_new.columns])
# Find out the number of missing values in each variables in the dataset
print("\n\nThe number of missing values in each column:")
hitters_new.isnull().sum(axis = 0)
```

```python
# Fill the missing values in the salary column with the column mean
#hitters_n = hitters_new.fillna(hitters_new.mean())
# Drop the columns with missing values
hitters_n = hitters_new.dropna()
# Reset the index
hitters_n = hitters_n.reset_index()
#hitters_n.isnull().sum()
hitters_n.shape
```

```python
# Separate the predictors and the target variable
X = hitters_n.iloc[:,:-1]
y = hitters_n.iloc[:,-1:]
```

## 1.1 Lasso Regression

```python
# The coefficients trajectories
alphas_l = np.logspace(-3, 3, 200)
# Set the "normalize" argument to "True" to normalize the predictors
clf_l = linear_model.Lasso(normalize = True, max_iter = 1e5)
coefs_l = []
for a in alphas_l:
    clf_l.set_params(alpha = a)
    clf_l.fit(X, y)
    coefs_l.append(clf_l.coef_)
    
ax = plt.gca()
ax.plot(alphas_l, coefs_l)
ax.set_xscale('log')
plt.xlabel('alpha')
plt.ylabel('coefficients')
plt.title('Lasso coefficients as a function of the regularization')
plt.show()
```

```python
# Find the last three predictors as the regularization increases
# As the plot shows, the number of predictors decrease to three somewhere after alpha becomes greater than 5.
# We can find the point where there are only three predictors by testing different alphas.
# When the alpha increases to 10, there are only three predictors left.
clf_lt = linear_model.Lasso(normalize = True, max_iter = 1e5, alpha = 10)
clf_lt.fit(X, y)
for col, coef in zip(hitters_n, clf_lt.coef_):
    print(col+':', coef)
```

#### The final three predictors that remain in the model are "Hits", "CRuns" and "CRBI".

### Select the optimal regularization penalty using 50-fold cross validation

```python
model = LassoCV(cv=50, normalize = True).fit(X, y.values.ravel())
# Display results
ymin, ymax = 0, 400000
plt.plot(model.alphas_, model.mse_path_, ':')
plt.plot(model.alphas_, model.mse_path_.mean(axis=-1), 'k',
         label='Average across the folds', linewidth=2)
plt.axvline(model.alpha_, linestyle='--', color='k',
            label='Optimal alpha: CV estimate')

plt.legend()

plt.xlabel('alpha')
plt.ylabel('Mean squared error')
plt.title('Mean squared error on each fold')
plt.ylim(ymin, ymax)
plt.show()
print("The optimal regularization penalty:", model.alpha_)
```

**The optimal regularization penalty is around 0.181.**

```python
# Find out the number of predictors in that model
# Print out the coefficients
print("The coefficients of the model with the optimal regularization penalty:")
for col, coef_l in zip(hitters_n.columns, model.coef_):
        print(col, coef_l)
```

**There are 12 predictors in the model with the optimal regularization penalty estimate by cross validation.**

**Predict salaries using Lasso regression:**

```python
# Predict salary of each player using the Lasso model with the optimal regularization and the original predictor data
SalaryPred_l = model.predict(X)
SalaryPred_l = pd.DataFrame(SalaryPred_l)
SalaryPred_l.rename(columns = {0:'SalaryPred'}, inplace = True)
hitters_pred_l = pd.concat([hitters_n, SalaryPred_l], axis=1)
hitters_pred_l[['Salary', 'SalaryPred']].head()
```

## 1.2 Ridge Regression

```python
# The coefficient trajectory
clf_r = Ridge(normalize = True)
coefs_r = []
alphas_r = np.logspace(-3, 3, 200)

# Train the model with different regularization penalties
for a in alphas_r:
    clf_r.set_params(alpha = a)
    clf_r.fit(X, y)
    coefs_r.append(clf_r.coef_[0])

# Display results
ax = plt.gca()
ax.plot(alphas_r, coefs_r)
ax.set_xscale('log')
plt.xlabel('alpha')
plt.ylabel('coefficients')
plt.title('Ridge coefficients as a function of the regularization')
plt.axis('tight')
plt.show()
```

```python
# Use cross validation to find out the optimal regularization value
# For simplicity, we use LOOCV, which is the default CV method of the RidgeCV function
## and only if using this method, we can take advantage of the function's ability to store mean squared errors
model_r = RidgeCV(alphas = alphas_r, normalize = True, cv = 50)
model_r.fit(X, y.values.ravel())
print("The optimal regularization penalty:", model_r.alpha_)
```

### The optimal regularization penalty is 1.275 and the estimated coefficients under this penalty are as follows.

```python
# Find out which predictors are in the model with the optimal regularization penalty
for col, coef_or in zip(hitters_n.columns, model_r.coef_):
    print(col, coef_or)
```

**Using Ridge regression and the optimal regularization estimated by 50-fold cross validation, the model predicting salaries contains all the predictors. This result is within expectation because Ridge regression does not shrink the coefficients to zero but only shrinks them to a very small magnitude.**

**Predict salaries using Ridge regression:**

```python
# Predict salary of each player using the Ridge model with the optimal regularization and the original predictor data
SalaryPred_r = model_r.predict(X)
SalaryPred_r = pd.DataFrame(SalaryPred_r)
SalaryPred_r.rename(columns = {0:'SalaryPred'}, inplace = True)
hitters_pred_r = pd.concat([hitters_n, SalaryPred_r], axis=1)
hitters_pred_r[['Salary', 'SalaryPred']].head()
```

# Question 2

The bias-variance tradeoff is the dilemma that predictive models with a lower bias tend to have a higher variance of the parameter estimates across different samples.Models with a higher variance usually have more predictors than models with a lower variance and tend to have more curvature. But models with less predictors have a higher bias.<p>
Regularization reduces the variance of the model by limiting the variation of the coefficients.<p>
In Question 1, the Lasso regression model with the optimal regularization have only 11 predictors while the Ridge regression with the optimal regularization have all of the 16 predictors. The Lasso model has a lower variance and a higher bias, whereas the Ridge regression has a higher variance and a lower bias.
