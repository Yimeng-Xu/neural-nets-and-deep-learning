```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def softmax(z):
    return np.exp(z)/(np.sum(np.exp(z), axis = 1).reshape(-1, 1))
def yhat(X, W1, b1, W2, b2, V, c):
    # returns model prediction 
    h1 = ff_nn_2_ReLu(np.dot(X,W1) + b1)
    h2 = ff_nn_2_ReLu(np.dot(h1, W2) + b2)
    y = softmax(np.dot(h2,V) + c)
    return y
```

```python
X = np.array([1, 0, 0, -1, -1, 1]).reshape(2, -1).T
W1 = np.array([1, 0, -1, 0, 0, 0.5]).reshape(3, -1).T
W2 = np.array([1, 0, 0, -1, -1, 0]).reshape(2, -1).T
V = np.array([1, 1, 0, 0, -1, -1]).reshape(3, -1).T
b1 = np.array([0,0,1])
b2 = np.array([1, -1])
c = np.array([1, 0, 0])
```

```python
# The final answers
# The class probabilities associated with the forward pass of each of the three samples
yhat(X, W1, b1, W2, b2, V, c)
```

#### Each row of the above matrix is the probabilities associated with a certain sample in X.

```python
##### Calculate the probabilities step by step ######
# a1, the input for hidden layer H1
a1 = np.dot(X, W1)+b1
# h1, the output of hidden layer H1
h1 = ff_nn_2_ReLu(a1)
# a2, the input for hidden layer H2
a2 = np.dot(h1, W2)+b2
# h2, the output of hidden layer H2
h2 = ff_nn_2_ReLu(a2)
# v, the input for the output function softmax
v = np.dot(h2, V)+c
# y, the required probabilities
y = softmax(v)
y
```

## (2)

```python
import numpy as np
import matplotlib.pyplot as plt

#Plot of Rosenbrock's banana function: f(x,y)=(1-x)^2+100(y-x^2)^2

rosenbrockfunction = lambda x,y: (1-x)**2+100*(y-x**2)**2

n = 100 # number of discretization points along the x-axis
m = 100 # number of discretization points along the y-axis
a=-3; b=3 # extreme points in the x-axis
c=-8; d=8 # extreme points in the y-axis

X,Y = np.meshgrid(np.linspace(a,b,n), np.linspace(c,d,m))

Z = rosenbrockfunction(X,Y)

plt.figure(figsize = (12,8))
plt.contour(X,Y,Z,np.logspace(-0.5,3.5,20,40, base=10))

plt.title('f(x,y)=(1-x)^2+100(y-x^2)^2')
plt.xlabel('x')
plt.ylabel('y')

plt.show()
```

## (3)

```python
def grad_f(vector):
    x, y = vector
    df_dx = 400*x**3-400*x*y+2*x-2
    df_dy = 200*y-200*x**2
    return np.array([df_dx, df_dy])
def gradient_descent(starting_point = None, learning_rate = 0.0005, iterations = 13000):
    if starting_point:
        point = starting_point
    else:
        point = np.array([2, 5])
    trajectory = [point]
    for i in range(iterations):
        grad = grad_f(point)
        point = point - learning_rate * grad
        #print('grad', grad)
        #print('point',point)
        trajectory.append(point)
    return(np.array(trajectory))
```

```python
np.random.seed(10)
traj1 = gradient_descent()
np.random.seed(10)
traj2 = gradient_descent(learning_rate = 0.001)
np.random.seed(10)
traj3 = gradient_descent(learning_rate = 0.0015)
np.random.seed(10)
traj4 = gradient_descent(learning_rate = 0.002)

fig, ax = plt.subplots(figsize = (12,8))
ax.set_xlim([-3,3])
ax.set_ylim([-8,8])
CS = ax.contour(X,Y,Z,np.logspace(-0.5,3.5,20,40, base=10))
plt.plot(traj1[:,0], traj1[:,1], '-o', color = '#3ACDFA', label = 'learning_rate=0.0005')
plt.plot(traj2[:,0], traj2[:,1], '-o', color = '#F1FF13', label = 'learning_rate=0.001')
plt.plot(traj3[:,0], traj3[:,1], '-o', color = '#FA0111', label = 'learning_rate=0.0015')
plt.plot(traj4[:,0], traj4[:,1], '-o', color = '#5ACE3C', label = 'learning_rate=0.002')
plt.plot(2, 5, "+", color = 'black', markersize = 20)
plt.legend(loc = 'lower left')
```

Our gredient descent algorithm starts at point [2,5] and makes 13000 iterations.<p>
Theoretically, when x=1 and y=1 the functions reaches its minimum.<p>
When learning rate is 0.0005, the algorithm walks toward the right minimum point but does not reach it because the learning rate is too low for an iteration of 13000. As the number of iterations increases, the algorithm will eventually find the minimum.<p>
When learning rate is 0.001, the gradient descent algorithm can walk very close to the right minimum.<p>
When the learning rate increases to 0.0015 or greater, the gradient descent algorithm no longer outputs the right minimum point. Instead, the trajectory oscillates randomly and in great steps. The reason lies in the property of the Rosenbrock function. Its gradient can easily become very large, which causes the gradient descent algorithm to walk in large steps and finally drifts far away from the minimum point.<p>

## (4)

```python
def grad_descent_with_momentum(starting_point=None, iterations=10, epsilon=0.0005, alpha = 0.7):
    if starting_point:
        point = starting_point
    else:
        point = np.array([np.random.uniform(-3,3,size=1)[0], np.random.uniform(-3,3,size=1)[0]])
    #print(point)
    trajectory = [point]
    v = np.zeros(point.size)
    for i in range(iterations):
        grad = grad_f(point)
        v = alpha*v + epsilon*grad
        point = point - v
        trajectory.append(point)
    return np.array(trajectory)
```

```python
np.random.seed(10)
traj1 = grad_descent_with_momentum(iterations=8000, epsilon=0.0005, alpha=0.6)
np.random.seed(10)
traj2 = grad_descent_with_momentum(iterations=8000, epsilon=0.0005, alpha=0.01)
np.random.seed(10)
traj3 = grad_descent_with_momentum(iterations=8000, epsilon=0.0005, alpha=1)
np.random.seed(10)
traj4 = grad_descent_with_momentum(iterations=8000, epsilon=0.0001, alpha=0.6)
np.random.seed(10)
traj5 = grad_descent_with_momentum(iterations=8000, epsilon=0.003, alpha=0.6)

fig, ax = plt.subplots(figsize = (12,8))
ax.set_xlim([-3,3])
ax.set_ylim([-8,8])
CS = ax.contour(X, Y, Z,np.logspace(-0.5,3.5,20,40, base=10))
plt.plot(traj1[:,0], traj1[:,1], '-o', color = '#31CDFA', label = 'eps=0.0005 alpha=0.6')
plt.plot(traj2[:,0], traj2[:,1], '-o', color = '#1233FF', label = 'eps=0.0005 alpha=0.01')
plt.plot(traj3[:,0], traj3[:,1], '-o', color = '#FA4511', label = 'eps=0.0005 alpha=1')
plt.plot(traj4[:,0], traj4[:,1], '-o', color = '#7ACE3C', label = 'eps=0.0001 alpha=0.6')
plt.plot(traj5[:,0], traj4[:,1], '-o', color = '#FFCECD', label = 'eps=0.003 alpha=0.6')
plt.plot(1.62792386,-2.8754883,'+', color = 'black', markersize = 20)
plt.legend(loc = 'lower left')
```

The black "+" indicates the starting point and we use an iteration of 8000.<p>
The value of alpha indicates how much the previous momentum weighs in the current momentum while the value of epsilon indicates how much the recently computed gradient weighs in the current momentum. If alpha is large, the algorithm tends to keep going in the previous direction even when it's time to change directions, as is illustrated by the red line in the graph. If epsilon is large, the algorithm tends to overreact according to the most recent gradient, which results in drastic oscillation as the pink line in the graph.

```python
def grad_f_w3(y,x3,w3):
    s3 = x3.dot(w3)
    grad_w3 = np.zeros_like(w3)
    delta3 = np.zeros_like(s3)
    
    for j in range(w3.shape[1]):
        grad_w3[:, j] = -(y[j]-np.exp(s3[j])/np.sum(np.exp(s3)))*x3
        delta3[j] = -(y[j]-np.exp(s3[j])/np.sum(np.exp(s3)))
    return grad_w3,delta3

def grad_relu(x):
    return np.where(x>0,1,0)
def relu(x):
    return np.where(x>0,x,0)
def grad_f_w2(delta3,x2,w2,w3):
    s2 = x2.dot(w2)
    grad_w2 = np.zeros_like(w2)
    delta2 = np.zeros_like(s2)
    
    for j in range(w2.shape[1]):
        grad_w2[:,j] = np.sum(delta3*w3[j+1,:])*grad_relu(s2[j])*x2
        delta2[j] = np.sum(delta3*w3[j+1,:])*grad_relu(s2[j])
    return grad_w2,delta2
```

```python
def grad_f(x,y,w1,w2,w3):
    x1 = np.append(1,x)
    x2 = np.append(1,relu(x1.dot(w1)))
    x3 = np.append(1,relu(x2.dot(w2)))
    grad_w3,delta3 = grad_f_w3(y,x3,w3)
    grad_w2,delta2 = grad_f_w2(delta3,x2,w2,w3)
    grad_w1,delta1 = grad_f_w2(delta2,x1,w1,w2)
    return (grad_w1,grad_w2,grad_w3)
```

```python
# Generate the synthetic dataset of 3 equally sampled bivariate Gaussian distributions
n_of_sample = 300
mu1 = [0, 2]
mu2 = [2,-2]
mu3 = [-2,-2]
cov = [[1, 0], [0, 1]]
y1 = np.array([[1,0,0]]*n_of_sample)
y2 = np.array([[0,1,0]]*n_of_sample)
y3 = np.array([[0,0,1]]*n_of_sample)
import matplotlib.pyplot as plt
x1 = np.random.multivariate_normal(mu1, cov, n_of_sample)
x2 = np.random.multivariate_normal(mu2, cov, n_of_sample)
x3 = np.random.multivariate_normal(mu3, cov, n_of_sample)
plt.plot(x1[:,0], x1[:,1], 'x')
plt.plot(x2[:,0], x2[:,1], 'o')
plt.plot(x3[:,0], x3[:,1], '.')
plt.show()
```

## (4)

```python
# Initialize the weights
w3 = np.array([1.,0,0,1,0,-1,1,0,-1]).reshape(3,-1)
w2 = np.array([1.,-1,1,-1,0,-1,0,0]).reshape(4,-1)
w1 = np.array([0,0,1.,1,-1,0,0,0,0.5]).reshape(3,-1)
```

```python
x = np.concatenate([x1,x2,x3])
y = np.concatenate([y1,y2,y3])
```

```python
def step(w1,w2,w3,a):
    """
    Return the parameters after each adjustment using gradient descent.
    """
    grad_w1_all = np.zeros_like(w1)
    grad_w2_all = np.zeros_like(w2)
    grad_w3_all = np.zeros_like(w3)
    
    for xi,yi in zip(x,y):
        (grad_w1,grad_w2,grad_w3) = grad_f(xi,yi,w1,w2,w3)
        grad_w1_all += grad_w1
        grad_w2_all += grad_w2
        grad_w3_all += grad_w3
    grad_w1_all = grad_w1_all/(3*n_of_sample)
    grad_w2_all = grad_w2_all/(3*n_of_sample)
    grad_w3_all = grad_w3_all/(3*n_of_sample)
    
    w1 = w1 - a*grad_w1_all
    w2 = w2 - a*grad_w2_all
    w3 = w3 - a*grad_w3_all
    return w1,w2,w3
```

```python
def pred(x,w1,w2,w3):
    """
    Return the softmax output
    """
    x1 = np.concatenate((np.ones(x.shape[0]).reshape(-1,1),x),1)
    x2 = relu(x1.dot(w1))
    x2 = np.concatenate((np.ones(x2.shape[0]).reshape(-1,1),x2),1)
    x3 = relu(x2.dot(w2))
    x3 = np.concatenate((np.ones(x2.shape[0]).reshape(-1,1),x3),1)
    s3 = x3.dot(w3)
    return np.exp(s3)/(np.sum(np.exp(s3), axis = 1).reshape(-1, 1))
```

```python
def loss(y,y_hat):
    return(np.mean(-np.sum(y*np.log(y_hat), axis = 1)))
```

```python
a = 0.1
l = [] # Store the loss of each step
for _ in range(500):
    w1,w2,w3 = step(w1,w2,w3,a)
    y_ = pred(x,w1,w2,w3)
    l.append(loss(y, y_))
    if _%10 == 0:
        print(_)
plt.plot(l)
plt.xlabel("Iteration")
plt.ylabel("Loss")
plt.show()
```

```python
y_ = pred(x,w1,w2,w3)
y_hat = np.argmax(y_,axis=1)

x1 = x[y_hat==0]
x2 = x[y_hat==1]
x3 = x[y_hat==2]
plt.plot(x1[:,0], x1[:,1], 'x')
plt.plot(x2[:,0], x2[:,1], 'o')
plt.plot(x3[:,0], x3[:,1], '.')
plt.show()
```

```python
def step(w1,w2,w3,a, b, v1, v2, v3):
    """
    Return the parameters after each adjustment using gradient descent.
    """
    grad_w1_all = np.zeros_like(w1)
    grad_w2_all = np.zeros_like(w2)
    grad_w3_all = np.zeros_like(w3)
    
    for xi,yi in zip(x,y):
        (grad_w1,grad_w2,grad_w3) = grad_f(xi,yi,w1,w2,w3)
        grad_w1_all += grad_w1
        grad_w2_all += grad_w2
        grad_w3_all += grad_w3
    grad_w1_all = grad_w1_all/(3*n_of_sample)
    grad_w2_all = grad_w2_all/(3*n_of_sample)
    grad_w3_all = grad_w3_all/(3*n_of_sample)
    
    v1 = b*v1 + a*grad_w1_all
    w1 = w1 - v1
    v2 = b*v2 + a*grad_w2_all
    w2 = w2 - v2
    v3 = b*v3 + a*grad_w3_all
    w3 = w3 - v3
    return w1,w2,w3, v1, v2, v3
```

```python
a = 0.1
b = 0.3
l_m = [] # Store the loss of each step
# Initialize momentum
v1 = np.zeros_like(w1)
v2 = np.zeros_like(w2)
v3 = np.zeros_like(w3)
# Initialize the weights
w3 = np.array([1.,0,0,1,0,-1,1,0,-1]).reshape(3,-1)
w2 = np.array([1.,-1,1,-1,0,-1,0,0]).reshape(4,-1)
w1 = np.array([0,0,1.,1,-1,0,0,0,0.5]).reshape(3,-1)
for _ in range(500):
    w1,w2,w3, v1, v2, v3 = step(w1,w2,w3,a, b, v1, v2, v3)
    y_ = pred(x,w1,w2,w3)
    l_m.append(loss(y, y_))
    if _%10 == 0:
        print(_)
```

```python
y_ = pred(x,w1,w2,w3)
y_hat = np.argmax(y_,axis=1)

x1 = x[y_hat==0]
x2 = x[y_hat==1]
x3 = x[y_hat==2]
plt.plot(x1[:,0], x1[:,1], 'x')
plt.plot(x2[:,0], x2[:,1], 'o')
plt.plot(x3[:,0], x3[:,1], '.')
plt.show()
```
